```python
from odc.geo import Geometry

from csdr.io import read_geospatial_file
from csdr.provenance import read_provenance
from csdr.utils import (
    get_geom_from_gdf,
    load_xarray_stacgeoparquet,
    open_stacgeoparquet,
    xarray_calculate_area,
)
```

```python
geometry_provenance_url = "s3://files.auspatious.com/csdr/geometries/EEZ_land_union_v4_202410.parquet.provenance.json"
dataset_provenance_url = (
    "s3://csdr-public-dev/datasets/gmw-v4/0-0-1/gmw.parquet.provenance.json"
)

# Load the provenance file
geometry_provenance = read_provenance(geometry_provenance_url)
geometry_file_url = geometry_provenance.get("dataUrl")

# Get the datset file
dataset_provenance = read_provenance(dataset_provenance_url)
dataset_file_url = dataset_provenance.get("dataUrl")
```

```python
gdf = read_geospatial_file(geometry_file_url)
geometry_id = "1643908b-6e6d-556f-ac60-226bed7d3b82"
geometry = gdf[gdf["csdr-id"] == geometry_id]
geometry.explore()
```

```python
odc_geom = Geometry(geometry.geometry.values[0], crs=geometry.crs)
```

```python
items = open_stacgeoparquet(dataset_file_url)

# Load the dataset
data = load_xarray_stacgeoparquet(
    items,
    geom=geometry,
    resolution=100,
    output_crs="epsg:6933",
    # measurements=[variable],
    chunks={},
)
data
```

```python
data = data.compute()
```

```python
data.mangrove.squeeze().plot.imshow()
```

```python
area = xarray_calculate_area(data, odc_geom, "mangrove", 1)
area
```

## Create acsc-ga-2015 geometries outputs

**Note** we need to use the same geometry output IDs in the product outputs


```python
geometries_parquet_path = "../geometries/acsc-ga-2015/out/acsc-primary-compartments.parquet"

# Read the parquet file
gdf = gpd.read_parquet(geometries_parquet_path)

# Print summary
print(f"\n{'='*50}")
print(f"Upload Summary:")
print(f"Total geometries in file: {len(gdf)}")

# print columns
print(f"Columns: {gdf.columns}")
```

```python
geometries_run_id = "acsc-ga-2015-run-1"

def get_geometry_name_and_id(row):
    # Get the name for this feature
    feature_name = (row['name'] if 'name' in row.index else f"Feature {idx}" )
    
    # Create unique ID using slugified name and index
    feature_id = f"{geometries_run_id}-{slugify(feature_name)}"

    return feature_name, feature_id

def slugify(text):
    """Convert text to a slug format suitable for IDs"""
    if text is None:
        return "unknown"
    # Convert to lowercase
    text = str(text).lower()
    # Replace spaces and underscores with hyphens
    text = re.sub(r'[\s_]+', '-', text)
    # Remove invalid characters
    text = re.sub(r'[^a-z0-9\-]', '', text)
    # Remove multiple consecutive hyphens
    text = re.sub(r'-+', '-', text)
    # Strip hyphens from start and end
    text = text.strip('-')
    return text if text else "unknown"

# Loop through each feature one at a time
for idx, row in gdf.iterrows():
      feature_name, feature_id = get_geometry_name_and_id(row)
      
      # # Validate and extract geometry
      geom = row.geometry
      
      # Check if it's a valid polygon or multipolygon
      if geom.geom_type not in ['Polygon', 'MultiPolygon']:
          print(f"⚠️  Row {idx}: Skipping - geometry type is {geom.geom_type}, not Polygon/MultiPolygon")
          continue
      
      # Convert geometry to GeoJSON format
      geometry_geojson = mapping(geom)

      # Create metadata dictionary with all columns except geometry
      metadata = {}
      for col in gdf.columns:
          if col != 'geometry':
              value = row[col]
              if isinstance(value, (int, float)):
                if not math.isnan(value):
                  metadata[col] = value
              elif isinstance(value, (str, bool)):
                  metadata[col] = value
      
      # Create the payload for this individual feature
      payload = {
          "geometriesRunId": "acsc-ga-2015-run-1",
          "id": feature_id,
          "name": feature_name,
          "description": f"Marine Regions EEZ geometry for {feature_name}",
          "metadata": {"some": "metadata"},
          "geometry": geometry_geojson,
          "properties": metadata  # Same as metadata for now
      }
      
      # Make the POST request for this single feature
      res = requests.post(
          "http://localhost:3000/api/v0/geometry-output",
          headers={
              "Content-Type": "application/json",
              "x-api-key": API_KEY
          },
          json=payload
      )
      
      # Check response
      if res.status_code in [200, 201]:
          print(f"✓ [{idx+1}/{len(gdf)}] Successfully uploaded: {feature_name} (ID: {feature_id})")
      else:
        print(f"Failed to upload: {feature_name} (ID: {feature_id})\nStatus: {res.status_code}, Response: {res.json()}")


print(f"Upload Complete!")
```

## Create product outputs

```python
# Loop through each class, and then post outputs for percent_cover, and total_area

variable_map = {
  2: {'percent_cover': 'tidal-flats-percentage', 'total_area': 'tidal-flats-area'},
  3: {'percent_cover': 'mangroves-percentage', 'total_area': 'mangroves-area'},
  4: {'percent_cover': 'saltmarsh-percentage', 'total_area': 'saltmarsh-area'},
  5: {'percent_cover': 'seagrass-percentage', 'total_area': 'seagrass-area'}
}

# Create geometryOutputId column using get_geometry_name_and_id
gdf['geometryOutputId'] = gdf.apply(lambda row: get_geometry_name_and_id(row)[1], axis=1)

def publish_class(gdf, class_id, col_name):
  variable_id = variable_map[class_id][col_name]

  # Filter our class 2
  filtered_df = gdf[gdf['class'] == class_id]

  # Create dataframe with just percent_cover and geometryOutputId (rename percent_cover to value)
  filtered_df = filtered_df[['geometryOutputId', col_name]].rename(columns={col_name: 'value'})

  # Convert to a JSON array (of objects with `geometryOutputId` and `value`)
  filtered_df_json = json.loads(filtered_df.to_json(orient='records'))

  print(f"Publishing {col_name} for class {class_id} ({variable_id}), {len(filtered_df_json)} outputs")
  

  # Post the data
  payload ={
        "productRunId": "australian-coastal-ecosystem-zonal-stats-run-1",
        "variableId": variable_id,
        "timePoint": "2021-01-01T00:00:00Z",
        "outputs": filtered_df_json
      }

  print(payload)

  res = requests.post(
      "http://localhost:3000/api/v0/product-output/bulk",
      headers={
        "Content-Type": "application/json",
        "x-api-key": API_KEY
      },
      json=payload
  )

  print(res.json())

publish_class(gdf, 2, 'percent_cover')
publish_class(gdf, 2, 'total_area')
publish_class(gdf, 3, 'percent_cover')
publish_class(gdf, 3, 'total_area')
publish_class(gdf, 4, 'percent_cover')
publish_class(gdf, 4, 'total_area')
publish_class(gdf, 5, 'percent_cover')
```

```python
import os
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import requests
from shapely.geometry import box
import zipfile
```

```python
asgs_ste_url = "https://www.abs.gov.au/statistics/standards/australian-statistical-geography-standard-asgs-edition-3/jul2021-jun2026/access-and-downloads/digital-boundary-files/STE_2021_AUST_SHP_GDA2020.zip"
asgs_ste_zip = "in/asgs-ste-2021.zip"
asgs_ste_dir = "in/asgs-ste-2021"
```

```python
if not os.path.exists("in/"):
    os.makedirs("in/", exist_ok=True)

if not os.path.exists("out/"):
    os.makedirs("out/", exist_ok=True)

print(f"Downloading ASGS STE2021 data...")
response = requests.get(asgs_ste_url)
with open(asgs_ste_zip, "wb") as f:
    print(f"Writing to {asgs_ste_zip}")
    f.write(response.content)
```

```python
print(f"Unzipping {asgs_ste_zip} to {asgs_ste_dir}")
with zipfile.ZipFile(asgs_ste_zip, "r") as zip_ref:
    zip_ref.extractall(asgs_ste_dir)
```

```python
shapefile_path = f"{asgs_ste_dir}/STE_2021_AUST_GDA2020.shp"
print(f"Reading {shapefile_path}")
gdf = gpd.read_file(shapefile_path)
print(f"Projecting to WGS84 from {gdf.crs}")
gdf = gdf.to_crs("EPSG:4326")

# Write out geoparquet
print(f"Writing to out/ste2021.parquet")
gdf.to_parquet("out/ste2021.parquet")
```
